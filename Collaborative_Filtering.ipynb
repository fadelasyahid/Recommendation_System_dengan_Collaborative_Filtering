```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
rating = pd.read_csv('ratings.csv')
film = pd.read_csv('movies.csv')
```

```python
rating.head(3)
```

```python
film.head(3)
```

```python
## melakukan merging

data = pd.merge(film,rating,how='inner',on='movieId')
```

```python
data
```

```python
## hilangkan timestamp 

data.drop('timestamp',axis=1,inplace=True)
```

```python
data
```

```python
## melihat salah satu contoh user ID dalam memberikan rating film

data[data['userId']==668]
```

```python
data_pivot = data.pivot_table(values='rating',index='userId',columns='title')
data_pivot
```

```python
data_pivot.replace(to_replace=np.NaN,value=0,inplace=True)
```

```python
data_pivot
```

```python
corr = data_pivot.corr()
```

```python
## misal user suka film 'Toy Story (1995)' dengan memberikan rating 5

user = ['Toy Story (1995)', 5]
```

```python
# similarity score berdasarkan correlation matrix

result = corr[user[0]] * user[1]
result
```

```python
result.min()
```

```python
result.max()
```

```python
## urutkan dari yang terbesar hingga terkecil


result.sort_values(ascending=False)
```

```python
## rekomendasi 5 film lain untuk user yang memberikan rating 5 pada film Toy Story (1995)

# result.sort_values(ascending=False)[:6]
for i in result.sort_values(ascending=False)[:6].index:
    print(f'film rekomendasi:',i)
```

```python
## bagaimana jika user memberikan rating kepada lebih dari 1 film?
## misal user suka Toy Story (1995) dengan memberkan rating 5 dan jumanji (1995) dengan rating 3

user1 = ['Toy Story (1995)', 5],['Jumanji (1995)', 3]
```

```python
for title, rating in user1:
    a = corr[title]*(rating/5) #-- supaya hasil nya 1 paling positif dan -1 paling negatif
    b = a.sort_values(ascending=False)
hasil = b
hasil
```

```python
## film rekomendasi berdasarkan rating user1 Toy Story(1995) : 5, dan Jumanji (1995):3

a = []
for i in hasil.index:
    a.append(i)
hasil_rek = a[:10]
```

```python
hasil_fix = pd.DataFrame(hasil_rek)
```
